```python
import pandas as pd
from functions import common_features_checker, feature_deletor

import warnings
warnings.filterwarnings('ignore')
```

```python
app_train_poly = pd.read_csv('../data/cleanedData/cleaned_df.csv')
new_bureau_df = pd.read_csv('../data/cleanedData/new_bureau_df.csv')
payment_and_previous_df = pd.read_csv('../data/cleanedData/payment_and_previous_df.csv')
cash_balance_df = pd.read_csv('../data/cleanedData/cash_balance_df.csv')
```

**app_train_poly & cash_balance_df**

```python
common_features_checker(app_train_poly, cash_balance_df)
```

```python
app_train_poly = app_train_poly.merge(cash_balance_df, on='SK_ID_CURR')
```

**app_train_poly & new_bureau_df**

```python
# check if there are common featues between the bureau and main datasets
common_features = common_features_checker(new_bureau_df, app_train_poly)
common_features
```

```python
common_features.remove('SK_ID_CURR')
new_bureau_df = feature_deletor(new_bureau_df, common_features)
```

```python
bureau_and_main_df = new_bureau_df.join(app_train_poly.set_index('SK_ID_CURR'), on='SK_ID_CURR', how='inner')

bureau_and_main_df
```

**bureau_and_main_df & payment_and_previous_df**

```python
common_features = common_features_checker(bureau_and_main_df, payment_and_previous_df)
common_features
```

```python
common_features.remove('SK_ID_CURR')
payment_and_previous_df = feature_deletor(payment_and_previous_df, common_features)
```

```python
main_bureau_payment_df = bureau_and_main_df.merge(payment_and_previous_df, on='SK_ID_CURR')
```

```python
main_bureau_payment_df
```

**main_bureau_payment_df & cash_balance_df**

```python
common_features = common_features_checker(main_bureau_payment_df, cash_balance_df)
common_features
```

```python
common_features.remove('SK_ID_CURR')
main_bureau_payment_df = feature_deletor(main_bureau_payment_df, common_features)
```

```python
all_df = main_bureau_payment_df.merge(cash_balance_df, on='SK_ID_CURR')
```

```python
all_df
```

```python
all_df.TARGET.value_counts(normalize=True)
```

```python
app_train_poly.columns = app_train_poly.columns.str.replace('[^a-zA-Z0-9_]', '_')
all_df.columns = all_df.columns.str.replace('[^a-zA-Z0-9_]', '_')
```

```python
path = '../data/cleanedData/'
all_df.to_csv(path + 'all_df.csv', index=False)
```
